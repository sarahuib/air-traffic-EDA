```python
# we find nulls in columns (carrier,carrier group,type,scheduled,charter,total)
# we will fill this null (categorical) data with mode
# numerical with mean
```

```python
df["carrier"] = df["carrier"].fillna(df["carrier"].mode()[0])
```

```python
df1["carrier"] = df1["carrier"].fillna(df1["carrier"].mode()[0])
```

```python
df["carriergroup"] = df["carriergroup"].fillna(df["carriergroup"].mode()[0])
df["type"] = df["type"].fillna(df["type"].mode()[0])
```

```python
df1["carriergroup"] = df1["carriergroup"].fillna(df1["carriergroup"].mode()[0])
df1["type"] = df1["type"].fillna(df1["type"].mode()[0])
```

```python
df["Scheduled "] = df["Scheduled"].fillna(0)
```

```python
df1["Scheduled"] = df1["Scheduled"].fillna(0)
```

```python
df=df.fillna(value=df["Charter"].mean())
df=df.fillna(value=df["Total"].mean())
```

```python
df1=df1.fillna(value=df["Charter"].mean())
df1=df1.fillna(value=df["Total"].mean())
```
